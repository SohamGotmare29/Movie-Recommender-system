```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L 
```

### Removing Space 

Example --  'Sharukh' 'Khan' , 'Salman' 'Khan' 
 - if our model tries to find 'Khan' it will get confused which to chose therfore we are removing **Space**

```python
def remove_space(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
```

## Stemimng 

```python
# ['loving','loved','love']
# ['love','love','love']
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps=PorterStemmer()
```

```python
ps.stem('loving')
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english') # removing stopwords 
    
```

```python
vector = cv.fit_transform(df['tags']).toarray()
```

## Calculating similarities using cosine_similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
similarity[1]
```

## "enumerate" it will give index to similarity vector 

- as we are doing sorting because we are finding similarity between top 5 movies 

```python
list(enumerate(similarity[1]))
```

```python
def recommend(movie):
    index = df[df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    
    
    for i in distances[1:6]:  # for 
        print(df.iloc[i[0]].title)
        
    
```
